### Diabetes Dataset

Realizado por el “National Institute of Diabetes and Digestive and Kidney” en Estados Unidos, y se ha realizado con un conjunto de mujeres mayores de 21 años descendientes de la etnia indígena “Pima”. El fichero que se encuentra en esta plataforma está en formato csv (valores separados por comas) y contiene las siguientes variables:

- Pregnancies. Representa el número de embarazos por los que ha pasado la mujer

- Glucose. Expresa el nivel de glucosa en sangre, en unidades de mg/dl

- Blood pressure. El nivel de presión sanguínea, en unidades de mmHg

- Skin thickness. El grosor de la piel expresado en mm

- Insulin. El nivel de insulina expresado en mg/dl

- BMI. El índice de masa corporal (IMC en castellano), que se calcula al dividir el peso, en kilogramos, entre el cuadrado de la estatura, en metros. Las unidades del BMI son kg/m2

- Diabetes pedigree function. Expresa el porcentaje de diabetes estimado, que se calcula dependiendo de la edad de la persona y su historial familiar.

- Age. Representa los años de las pacientes.

- Outcome. Representa el resultado final del análisis, en valores de 1 si el resultado final es que la paciente tiene diabetes, o 0 si el resultado es que no tiene diabetes. Es la variable dependiente.able dependiente.

````python
df = pd.read_csv("diabetes.csv", sep=",")
df.head()
````

### EJERCICIO 14

````python
df["Age_ranges"] = pd.cut(df["Age"], bins=[0,30,50,100], labels=["<30", "30-50", ">50"])
df1 = df.pivot_table("Insulin", columns="Outcome", index="Age_ranges", aggfunc="count")
df1
````

````python
fig, ax = plt.subplots(figsize=(7,4))
ax.bar(df1.index, 
        df1[0],
        label="Out0")
ax.bar(df1.index,
        df1[1],
        label="Out1",
        bottom=df1[0])
ax.legend()
````
