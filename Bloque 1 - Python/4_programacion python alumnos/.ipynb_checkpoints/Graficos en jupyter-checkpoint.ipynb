```python
import matplotlib.pyplot as plt

def pintar_grafico(datos, tipo_grafico='barras'):
    """
    Pinta un gráfico de barras o de líneas a partir de datos en un diccionario.

    :param datos: Un diccionario donde las claves son las etiquetas y los valores son listas de datos.
    :param tipo_grafico: Tipo de gráfico ('barras' o 'lineas'). Por defecto, 'barras'.
    """
    if tipo_grafico not in ['barras', 'lineas']:
        raise ValueError("El tipo de gráfico debe ser 'barras' o 'lineas'.")

    etiquetas = list(datos.keys())
    valores = list(datos.values())

    if tipo_grafico == 'barras':
        for i, etiqueta in enumerate(etiquetas):
            plt.bar(etiqueta, sum(valores[i]), label=etiqueta)

        plt.xlabel('Categorías')
        plt.ylabel('Valor')
        plt.title('Gráfico de Barras')
        plt.legend()
        plt.show()
    else:  # tipo_grafico == 'lineas'
        for i, etiqueta in enumerate(etiquetas):
            plt.plot(valores[i], label=etiqueta)

        plt.xlabel('Puntos')
        plt.ylabel('Valor')
        plt.title('Gráfico de Líneas')
        plt.legend()
        plt.show()

# Ejemplo de uso
datos = {
    'Serie A': [10, 20, 15, 30, 25],
    'Serie B': [5, 15, 10, 25, 20],
    'Serie C': [15, 25, 20, 35, 30]
}

# Grafico de barras
pintar_grafico(datos, tipo_grafico='barras')

# Grafico de lineas
pintar_grafico(datos, tipo_grafico='lineas')
```

```python
import matplotlib.pyplot as plt

def pintar_grafico(datos_x, datos_y, colores, tooltips=None, title_x='Datos en el Eje X', title_y='Datos en el Eje Y', title='Gráfico Personalizado'):
    """
    Crea un gráfico personalizado con datos en el eje X y en el eje Y, colores personalizados y tooltips.

    :param datos_x: Un diccionario donde las claves son las etiquetas y los valores son listas de datos en el eje X.
    :param datos_y: Un diccionario donde las claves son las etiquetas y los valores son listas de datos en el eje Y.
    :param colores: Un diccionario que asigna colores a cada serie.
    :param tooltips: Un diccionario que asigna tooltips a cada punto de datos.
    """
    fig, ax = plt.subplots()

    for etiqueta in datos_x.keys():
        if etiqueta in colores:
            color = colores[etiqueta]
        else:
            color = 'blue'  # Color predeterminado si no se especifica

        if etiqueta in tooltips:
            texto_tooltip = tooltips[etiqueta]
        else:
            texto_tooltip = None

        ax.plot(datos_x[etiqueta], datos_y[etiqueta], label=etiqueta, color=color, marker='o', linestyle='-')

        if texto_tooltip:
            for x, y, tooltip in zip(datos_x[etiqueta], datos_y[etiqueta], texto_tooltip):
                ax.annotate(f'{tooltip}', (x, y), textcoords="offset points", xytext=(0, 10), ha='center')

    # Personalización del gráfico
    ax.set_xlabel(title_x)
    ax.set_ylabel(title_y)
    ax.set_title(title)
    #ax.set_xticklabels(ax.get_xticks(), rotation=45)
    ax.legend()
    plt.grid(True)
    plt.xticks(rotation=45)

    plt.show()

# Ejemplo de uso
datos_x = {
    'Serie A': [1, 2, 3, 4, 5],
    'Serie B': [1, 2, 3, 4, 5],
    'Serie C': [1, 2, 3, 4, 5]
}

datos_y = {
    'Serie A': [10, 20, 15, 30, 25],
    'Serie B': [5, 15, 10, 25, 20],
    'Serie C': [15, 25, 20, 35, 30]
}

colores = {
    'Serie A': 'red',
    'Serie B': 'green',
    'Serie C': 'blue'
}

tooltips = {
    'Serie A': ['A1', 'A2', 'A3', 'A4', 'A5'],
    'Serie B': ['B1', 'B2', 'B3', 'B4', 'B5'],
    'Serie C': ['C1', 'C2', 'C3', 'C4', 'C5']
}

pintar_grafico(datos_x, datos_y, colores, tooltips)
```
