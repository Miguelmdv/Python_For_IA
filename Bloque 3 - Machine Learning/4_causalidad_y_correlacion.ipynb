```python
from scipy.stats import spearmanr
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.random.uniform(0, 3, 500)
y = np.random.uniform(0, 1, 500)
plt.scatter(x, y);
```

```python
np.corrcoef(x, y)[0,1]
```

```python
spearmanr(x, y).correlation
```

- No est√°n correlacionadas
- No pueden ser causales

```python
x = np.random.uniform(0, 3, 300)
y = 6 * x + 1 + np.random.randn(300)
plt.scatter(x, y);
```

```python
np.corrcoef(x, y)[0,1]
```

```python
spearmanr(x, y).correlation
```

- Est√°n correlacionadas fuertemente y positivamente
- No podemos indicar causalidad, podr√≠a ser una relaci√≥n espuria (derivada de terceros factores no representados)

```python
x = np.random.uniform(0, 3, 300)
y = - 3  * x + 1 + 3 * np.random.randn(300)
plt.scatter(x, y);
```

```python
np.corrcoef(x, y)[0,1]
```

```python
spearmanr(x, y).correlation
```

- Correlaci√≥n negativa d√©bil
- No podemos indicar causalidad, podr√≠a ser una relaci√≥n espuria (derivada de terceros factores no representados)

- **CAUSALIDAD IMPLICA CORRELACI√ìN**
- **CORRELACION NO IMPLICA CAUSALIDAD** (tomarlo con pizas)

Como causalidad implica correlaci√≥n, **si no hay correlaci√≥n, no hay causalidad**

```python
x = np.random.uniform(0, 3, 300)
y = np.exp(2 * x) + 1 + 2 * np.random.randn(300)
plt.scatter(x, y);
```

```python
np.corrcoef(x, y)[0,1]
```

```python
spearmanr(x, y).correlation
```
